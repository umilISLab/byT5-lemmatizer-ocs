```python
import conllu
import csv

from config import data_root
```

```python
dataset = "train"
# dataset = "valid"
# dataset = "test"
```

```python
with open(f"{data_root}/chu_{dataset}.conllu") as fin:
    data = conllu.parse(fin.read())
data
```

```python
data[0][0]
```

```python
transformed = []
for sent in data:
    for t in sent:
        transformed += [[f': {t["form"].lower()}:{t["upos"]}', t["lemma"]]]

with open(f"{data_root}/chu_{dataset}.csv", "w") as fout:
    wrt = csv.writer(fout)
    wrt.writerows(transformed)
```
